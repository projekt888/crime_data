```python
import pandas as pd
import numpy as np

#read in Crime_statistics
file_to_load = "COBRA-2009-2018.csv"

# Read Purchasing File and store into Pandas data frame
crime_data = pd.read_csv(file_to_load)

crime_data.head()
```

```python
x = crime_data["Neighborhood"].value_counts().rename_axis('Neighborhood').to_frame('Crimes')
```
