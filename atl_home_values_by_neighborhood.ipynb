1. We use the crime database to extract unique neighborhood values.
2. There's a website called "niche" which has a url pattern for neighborhoods which is easy to guess.  Most of the time we guess OK.  We download the page.
3. Using Beautiful Soup we extract the house value from that page.
4. We put it into a dataframe.  Then write it out to CSV.
5. There's some manual spot checking before we hand that CSV back to the analysis.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os

crime_file = os.path.join("Resources", "COBRA-2009-2018.csv")
crime_df = pd.read_csv(crime_file)

neighborhoods = pd.DataFrame({'Neighborhood':crime_df['Neighborhood'].unique()})

neighborhoods.head()
```

```python
# Building NICHE urls.
dash = neighborhoods['Neighborhood'].tolist()
nicheprefix = "https://www.niche.com/places-to-live/n/"
nichesuffix = "-atlanta-ga/"
index = 0
for d in dash:
    d = str(d)
    d = d.lower()
    d = d.replace(" ","-")
    d = d.replace("---","-")
    d = d.replace("/","-")
    d = d.replace(".","")
    d = nicheprefix + d + nichesuffix
    dash[index] = d
    index += 1
    #print (d)
    
neighborhoods['Niche URL'] = dash
```

```python
def get_link(url):
    ret_value = 0
    from urllib.request import Request, urlopen
    from bs4 import BeautifulSoup as soup
    print(url)
    try:
        req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
        webpage=urlopen(req).read()
        page_soup = soup(webpage, "html.parser")

        mydivs = page_soup.findAll("div", {"class": "scalar__value"})
        
        #print (len(mydivs))
        #print (mydivs[1])
        ret_value = (mydivs[1].find("span").get_text())

    except:
        print ("Could Not Find")
        
    return ret_value

niche_price = []

niche_url = neighborhoods['Niche URL'].tolist()

for niche in niche_url:
    niche_price.append(get_link(niche))
```

```python
niche_price
```

```python
neighborhoods['Niche Price'] = niche_price
neighborhoods.head()

neighborhoods.to_csv("niche.csv", index=False, header=True)
```
